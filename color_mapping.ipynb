```python
from re import S
import matplotlib.pyplot as plt
import cv2
import numpy as np
import os

from load_exr import read_exr_as_np

def read_exr_as_stokes(fn):
    image, _ = read_exr_as_np(fn)   # image shape = [512, 512, 16]
    stokes = np.clip(image[..., 4:15:3], -1, 1)     # stokes shape = [512, 512, 4]
    
    return stokes
```

# RGB visualization for Stokes images at a viewpoint

```python
import pandas as pd

wave_to_rgb = pd.read_csv("spec_to_rgb.csv")

def color_mapping(PATH, FIG_OUTPATH, OBJECT, wavelength, FILE, scalar=1, s123_min=-1, s123_max=1):
    img_path = f"{PATH}/{OBJECT}/{wavelength}/train/{FILE}"
    
    stokes = read_exr_as_stokes(img_path)    
    H, W = stokes.shape[:2]
    rgb = np.array(wave_to_rgb[wave_to_rgb["wavelength"] == wavelength][["R", "G", "B"]])/255

    plt.figure()
    for i in range(4):
        s = np.tile(stokes[..., i][..., np.newaxis], (1, 3))       # s shape = [512, 512, 3]
        if i > 1:
            s = (s - s123_min)/(s123_max-s123_min)
        s = s * np.expand_dims(rgb,(0)) * scalar
        plt.subplot(1,4,i+1)
        plt.imshow(s)
        FILE_OUT = f"{FILE[:-4]}_s{i}_rgb_vis_{wavelength}.png"
        out_img_path = f"{FIG_OUTPATH}/{FILE_OUT}"
        cv2.imwrite(out_img_path, s[:,:,::-1]*255)
    plt.show()
            
```

```python
PATH = '//jarvis.postech.ac.kr/workspace/youngchan/HpNeRF/data/synthetic'    # directly accessing remote notework seems to take long time. better to use a local path 
wvls = np.arange(450, 650+1, 10)
for wvl in wvls:
    print(f'processing wavelength {wvl}')
    color_mapping(PATH, FIG_OUTPATH, "hotdog", wvl, "r_1.exr")
```

# Polarimetric-spectral plotting for a few spatial points

```python
PATH = '//jarvis.postech.ac.kr/workspace/youngchan/HpNeRF/data/synthetic'  # directly accessing remote notework seems to take long time. better to use a local path 
OBJECT = "hotdog"
FILE = "r_0.exr"
wvls = np.arange(450, 650+1, 10)

hp_stokes = []
for wvl in wvls:
    print(f'reading image @ wavelength {wvl}')
    stokes = read_exr_as_stokes( f"{PATH}/{OBJECT}/{wvl}/train/{FILE}")
    hp_stokes.append(stokes)

hp_stokes = np.array(hp_stokes)
```

```python
r, c = 300, 150
plt.figure()
plt.imshow(hp_stokes[0,:,:,0])
plt.plot(r,c,'*', c='r')
plt.title('location')
plt.colorbar()


fig, ax1 = plt.subplots()
color_s0='tab:red'
ax1.set_xlabel('Wavelength [nm]')
ax1.set_ylabel('${s}_0$')
ax1.plot(wvls, hp_stokes[:,r,c,0], '.-', color=color_s0)
ax1.tick_params(axis ='y', labelcolor = color_s0)

# Adding Twin Axis
color_s123='tab:green'
ax2 = ax1.twinx()
ax2.set_ylabel('${s}_{1,2,3}$')
ax2.plot(wvls, hp_stokes[:,r,c,1], '.--', color=color_s123)
ax2.plot(wvls, hp_stokes[:,r,c,2], '.-', color=color_s123)
ax2.plot(wvls, hp_stokes[:,r,c,3], '.:', color=color_s123)
ax2.tick_params(axis ='y', labelcolor = color_s123)

plt.savefig(f"{FIG_OUTPATH}/hp_stokes.svg")
```
